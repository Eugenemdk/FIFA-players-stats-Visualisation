```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
my_filepath='../input/fifa-player-stats-database/FIFA22_official_data.csv'
my_data=pd.read_csv(my_filepath)
```

```python
sns.lineplot(data=my_data.loc[my_data['Age']<19],x='Overall',y='Potential',hue='Age',style='Age')
#sns.lineplot(data=my_data,x='Overall',y='DefensiveAwareness')
```

```python
my_data.tail(20)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Nationality',y='Overall',data=my_data.iloc[1:20])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Best Overall Rating',y='Name',data=my_data.iloc[1:10])
```

```python

#g=sns.catplot(y='Overall',x='Name',kind='count',data=my_data.iloc[1:5],hue='Best Position')
my_pie=my_data.groupby('Best Position').size()
my_pie.plot(kind='pie',subplots=True, figsize=(8, 8),ylabel='Best Position of Players')
```

```python
sns.scatterplot(x=my_data['DefensiveAwareness'],y=my_data['Overall'],hue=my_data['Age'])
```

```python
sns.scatterplot(x='DefensiveAwareness',y='Overall',data=my_data.iloc[1:30],hue='Age',size='Age',sizes=(20, 400))
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Club',data=my_data.iloc[[23,29,1200,10,100,12]])
```

```python
sns.jointplot(data=my_data.iloc[1:20],x='Overall',y='Nationality')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=my_data.iloc[1:15],y='DefensiveAwareness',x='Potential')
```

```python
plt.figure(figsize=(12,6))
couple_columns=my_data[['Potential','Overall','DefensiveAwareness']]
couple_columns.head()
overall_n_potential=couple_columns.groupby(['Potential','Overall','DefensiveAwareness']).mean()
print(overall_n_potential.shape)
overall_n_potential.head(10)
overall_n_potential=overall_n_potential.reset_index()
overall_n_potential.head()
overall_n_potential_table=overall_n_potential.pivot_table('Potential','Overall','DefensiveAwareness').fillna(0).head(5)
sns.heatmap(overall_n_potential_table,annot=False)
```

```python
sns.lmplot(x="Potential", y="Overall", hue="Age", data=my_data.loc[my_data['Age']<19])
```
